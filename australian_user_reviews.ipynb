```python
import pandas as pd
import ast

#abrir y convertir en DataFrame
rows = []
with open('C:/Users/Susana y Juan/OneDrive/Escritorio/Henry/PI_ML_OPS/Archivos/australian_user_reviews.json', encoding = 'MacRoman') as users_reviews:
    for line in users_reviews.readlines():
        rows.append(ast.literal_eval(line))
        
review_df = pd.DataFrame(rows)
```

```python
#DESANIDAR COLUMNA

# Crear una lista para almacenar los datos desanidados
data_desanidada = []

# Iterar a través de las filas del DataFrame original
for index, row in review_df.iterrows():
    user_id = row['user_id']
    reviews = row['reviews']
    
    for i in reviews:
        new_row = {
            'user_id': user_id,
            'funny': i.get('funny', ''),
            'posted': i.get('posted', ''),
            'last_edited': i.get('last_edited', ''),
            'item_id': i.get('item_id', ''),
            'helpful': i.get('helpful', ''),
            'recommend': i.get('recommend', False),  # Cambié bool a False para que sea un valor booleano por defecto
            'review': i.get('review', '')
        }
        data_desanidada.append(new_row)

# Crear un nuevo DataFrame con los datos desanidados
desanidado_df = pd.DataFrame(data_desanidada)

# Ver el nuevo DataFrame
print(desanidado_df.head())  # Esto mostrará las primeras filas del DataFrame desanidado
```

```python
#UNIR DOS DATAFRAME (merge)
# Suponiendo que ya tienes los DataFrames review_df y desanidado_df

# Unir los DataFrames utilizando la columna "user_id" como clave
merged_df = review_df.merge(desanidado_df, on='user_id', how='inner')

# "how='inner'" significa que solo se incluirán las filas que tengan coincidencias en ambas tablas.
# Si deseas incluir todas las filas de ambas tablas, puedes usar "how='outer'" en su lugar.

# Ahora tienes un nuevo DataFrame llamado merged_df que contiene la información combinada de ambos DataFrames
merged_df
```

```python
#ELIMINAS COLUMNAS

# Supongamos que quieres eliminar las columnas 'columna1' y 'columna2' de tu DataFrame
columnas_a_eliminar = ['user_url', 'reviews', 'funny', 'last_edited']

# Utiliza el método drop para eliminar las columnas
df_reviews_clean = merged_df.drop(columnas_a_eliminar, axis=1)

# El argumento "axis=1" indica que quieres eliminar columnas en lugar de filas.
```

```python
#ELIMINAR PARTE DE UN REGISTRO

# Eliminar la palabra "Posted" de la columna "posted"
df_reviews_clean['posted'] = df_reviews_clean['posted'].str.replace('Posted', '')

# Ahora, la columna "mi_columna" tendrá la palabra "eliminar" eliminada en todas las filas.

df_reviews_clean.head()
```

```python
#TRANSFORMAR TIPOS DE DATOS 
df_reviews_clean['posted'] = df_reviews_clean['posted'].astype(str)
```

```python
# CHEQUIAR TIPO DE DATOS

tipo_de_dato = df_reviews_clean['posted'].dtypes
print(tipo_de_dato)
```

```python
# REEMPLAZAR valores None o NaN con cadenas vacías
df_reviews_clean['posted'] = df_reviews_clean['posted'].fillna('').astype(str)
```

```python
#ELIMINAR POSIBLES ESPACIOS EN BLANCO

# Eliminar espacios en blanco al principio de la columna 'Columna1'
df_reviews_clean['posted'] = df_reviews_clean['posted'].str.lstrip()
```

```python
# ELIMINAR PARTE DE UN REGISTRO

# Eliminar puntos al final de la columna 'Columna1'
df_reviews_clean['posted'] = df_reviews_clean['posted'].str.rstrip('.')
```

```python
#CREACIÓN NUEVA COLUMNA PARA AGREGAR INFORMACIÓN LUEGO

# Crear una nueva columna 'Date_Review' con valores nulos (NaN)
df_reviews_clean['Date_Review'] = pd.NaT

# Obtener el índice de la columna 'posted'
indice_columna_posted = df_reviews_clean.columns.get_loc('posted')

# Insertar la columna 'Date_Review' al lado de 'posted'
df_reviews_clean.insert(indice_columna_posted + 1, 'Date_Review', df_reviews_clean.pop('Date_Review'))
```

```python
#SE REALIZÓ PARA EXTRAER SÓLO EL AÑO. (formato de la fecha inválida para transformarla directamente)
# Extraer los últimos 4 caracteres de la columna 'posted' y asignarlos a 'Date_Review'
df_reviews_clean['Date_Review'] = df_reviews_clean['posted'].str[-4:]

# Encuentra los registros que comienzan con una letra en la columna 'Date_Review' (porque algunos archivos que no 
# contenían el año copiaban parte del mes. Asi que esos los quería eliminar ya que no tenían
# el dato que necesitaba)

registros_con_letras = df_reviews_clean[df_reviews_clean['Date_Review'].str[0].str.isalpha()]

# Cuenta cuántos registros cumplen con el criterio
cantidad_registros_con_letras = len(registros_con_letras)

print(cantidad_registros_con_letras)
```

```python
#AGREGA 'null' A LOS REGISTROS VACIOS

# Reemplaza los valores en esos registros con 'null'
df_reviews_clean.loc[df_reviews_clean['Date_Review'].str[0].str.isalpha(), 'Date_Review'] = 'null'

# Verifica que los valores se hayan reemplazado correctamente
print("Cantidad de registros con letras:", cantidad_registros_con_letras)
```

```python
# Convierte las columnas que contienen años en números enteros
df_reviews_clean['Date_Review'] = df_reviews_clean['Date_Review'].apply(lambda x: int(x) if x.isdigit() else x)
```

```python
#Eliminar columna 'posted'
df_reviews_clean = df_reviews_clean.drop(columns=['posted'])
```

```python
#VER CANTIDAD DE ARCHIVOS VACIOS POR COLUMNA

registros_vacios_por_columna = df_reviews_clean.isnull().sum()

# Mostrar el resultado
print(registros_vacios_por_columna)
```

```python
#PORCENTAJE DE NULOS EN UNA COLUMNA

# Reemplaza 'Nombre_Columna' con el nombre de la columna que deseas analizar
nombre_columna = 'Date_Review'

# Calcular el porcentaje de celdas que no están vacías y contienen 'null'
porcentaje_con_null = ((df_reviews_clean['Date_Review'].notnull()) & (df_reviews_clean['Date_Review'] == 'null')).sum() / len(df_reviews_clean) * 100

# Mostrar el porcentaje
print(f'Porcentaje de celdas que no están vacías y contienen "null" en la columna {nombre_columna}: {porcentaje_con_null:.2f}%')
```

```python
#ELIMINAR COLUMNA

# Supongamos que tienes un DataFrame llamado df y quieres eliminar la columna 'columna_a_eliminar'
df_reviews_clean = df_reviews_clean.drop('helpful', axis=1)
```

```python
# GUARDAR EN CSV

df_reviews_clean.to_csv('review_data.csv', index=False)  
# El argumento index=False evita que se guarde el índice en el archivo
```
